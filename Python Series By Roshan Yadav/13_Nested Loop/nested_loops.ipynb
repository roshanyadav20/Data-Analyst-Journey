Nested Loops in Python

- Loops inside another loop is known as nested loop.

Why do we use Nested Loops?

- Nested loops are used to handle multi-dimensional data, complex iterations, and pattern generation.


```python
# Print number from 1 to 3

for num in range(1,4):
    print(num)  
```

```python
# print numbers from 1 to 3 for 3 times.

for num in range(1,4):
    print(num)
    print("- - -")  
    
for num in range(1,4):
    print(num) 
    print("- - -")  
    
for num in range(1,4):
    print(num)
    print("- - -")           
```

“We use nested loops to make complex tasks easier.”

```python
for i in range(3):
    for num in range(1,4):
        print(num)
    print("- - -")    
```

```python
for i in range(3):
    print("Outer loop iteration no, ", i)
    for num in range(1,4):
        print(num)
       
```

While Loop

```python
# print 1 to 3 for 3 times using while loop
i = 1
while i < 4:
    print("while loop iteration no.", i)
    for j in range(1,4):       
        print(j) 
    i +=1   
        
```

```python
# print prime numbers between range of 2 to 10 using nested loop:

for i in range(2,10):
    for j in range(2,i):
        if i % j == 0:
            break
    else:
        print(i)
```

```python
for i in range(2,10):
    for j in range(2,i):
        if i % j == 0:
            break
    else:
        print(i)
```
