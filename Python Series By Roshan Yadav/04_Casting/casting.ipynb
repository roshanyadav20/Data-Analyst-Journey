Type Casting in Python

```python
a = 1
print(type(a))
```

```python
# b = "1"
# print(a+b) # error
```

```python
b = "1"
print(type(b))

c = int(b)
print(c)
print(type(c))

print(a+c)
```

```python
print(a+int(b))
```

```python
# all numerical type can be cast into str

mynum = 26
mynum2 = str(mynum)
print(type(mynum2))
```

```python
f1 = 22.56
print(type(f1))
print(type(int(f1)))
```

```python
f1 = 22.56
f2 = int(f1)
print(f2)
print(type(f2))
```

Implicit Type Casting

```python
# Integer to float convert:
var1 = 10
var2 = 15.5
var3 = var1 + var2
print(var3)
print(type(var3))
```

Explicit Type Castng 

```python
# converting string to integer:
int_num = 101
str_num = str(int_num)
print(type(str_num))
```

```python
a0 = bool(0)
print(a0)
print(type(a0))
```

```python
a1 = bool(1)
print(a1)
print(type(a1))
```
