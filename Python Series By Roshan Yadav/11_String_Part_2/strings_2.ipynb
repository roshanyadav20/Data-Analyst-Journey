2. String Slicing 

Syntax: string[start : end : step]

- start: The index to start slicing (inclusive). Default value is 0.
- end: The index to stop slicing (exclusive). Default value is length of string.
- step: How much to increment the index after each character. Default value is 1.

```python
my_name = "Roshan"
print(my_name[0:3])

print(my_name[0:3:1])

print(my_name[0:5:1])

print(my_name[3:5:1])

print(my_name[0:5:2]) # step 2

print(my_name[0:5:3]) # step 3

print(my_name[0:5:4]) # step 4

print(my_name[0:3:1])  # first 2 char

print(my_name[0:3]) # first 3 char

print(my_name[2:5]) # third to fifth char

print(my_name[1:4]) # second to forth char

print(my_name[-1:]) # last char of str

print(my_name[5:]) # last char of str

print(my_name[-2:]) # last 2 char of str

print(my_name[-3:]) # last 3 char of str

print(my_name[0::2]) # every second char

print(my_name[:]) # all char

print(my_name[::]) # all char

print(my_name[::-1]) # reverse the string

```

3. String Methods

```python
word = "Roshan, Yadav"

#1. len()
print(len(word))

#2. upper()
print(word)

#3. lower()
print(word.lower())

#4. count()
print(len(word))

#5. find()
print(word.find('s'))

#6. split()
print(word.split(','))
print(word.split())

#7. replace()
print(word.replace("Roshan", "Keshav"))

#8. title()
print(word.title()) # every first letter capital ho jayega.

#9. strip()
print(word.strip())

#9. strip()
word2 = " Roshan Yadav  "
print(len(word2))
print(word2.strip())

#10. join()
zwords = ("Roshan", "is", "Great")
print(" ".join(zwords))
print("-".join(zwords))


```
