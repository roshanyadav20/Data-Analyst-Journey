Tuples in Python

```python
# create tuple using paranthesis
my_tuple = (1,2,3)
print(my_tuple)
print(type(my_tuple))
```

```python
my_tuple1 = (1, "Madhav", True, 3.15)
print(my_tuple1)
```

```python
# without paraenthesis 

tuple2 = 1,2,3
print(tuple2)
print(type(tuple2))
```

```python
# tuple constructor
tuple2 = tuple((1,5,9))
print(type(tuple2))
print(tuple2)
```

```python
list1 = [1,2,3]
new_tuple = tuple(list1)
print(new_tuple)
```

```python
# create a single element
a = ("a",) # comma add
print(type(a))
```

```python
# acess tuple using indexing 
fruits = ('apple', 'mango', 'cherry')
print(fruits[0])
print(fruits[-1])
```

```python
# tuple slicing 
new_tuple = (10, 20, 30, 40, 50)
print(new_tuple[0:3:1])
print(new_tuple[0::2])
print(new_tuple[0::3])
```

Tuple iteration

```python
# for loop
fruits = ('apple', 'mango', 'cherry')
for i in fruits:
    print(i)
```

```python
# while loop 
fruits1 = ('apple', 'mango', 'cherry')
i = 0
while i < len(fruits1):
    print(fruits1[i])
    i += 1
```

Tuple Methods

```python
color = ('blue', 'green', 'orange', 'blue')

# count
print(color.count('green'))

# index
print(color.index('blue'))
```

Packing and Unpacking Tuples:

- Packing is the process of putting multiple values into a single tuple.

- Unpacking is extracting values from a tuple into separate variables.

```python
a = "Madhav"
b = 21
c = "Engineering"

tuple_pack = a,b,c
print(tuple_pack)
```

```python
name, age, profession = tuple_pack
print("Name is",name)
print(age)
print(profession)
```
