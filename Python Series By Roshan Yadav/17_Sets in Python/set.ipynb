```python
# create set using curly braces
my_set = {1,2,3}
print(my_set)
print(type(my_set))
```

```python
# create set using set constructor
my_set2 = set([4,5,6])
print(my_set2)
```

```python
# set operations

# adding element
numbers = {1,2,3,4}
numbers.add(100)
print(numbers)
```

```python
# remove elements

fruits = {'apple', 'mango', 'banana'}
fruits.remove('banana')
print(fruits)
```

```python
#discard
fruits.discard('cherry') # does not show error
print(fruits)
```

Set Methods

```python
# Union - combine elemnts from two set
set_a = {1,2,3,4}
set_b = {3,4,5}
union_set = set_a.union(set_b)
print(union_set)
```

```python
# union alternative
union_set2 = set_a | set_b
print(union_set2)
```

```python
# Intersection - common elements
set1 = {1,2,3,4}
set2 = {3,4,5}
union_set = set1.intersection(set2)
print(union_set)
```

```python
# intersection alternarive
inter_set2 = set1 & set2
print(inter_set2)
```

```python
# Difference - element present in first set only but not in second set
set1 = {1,2,3,4}
set2 = {3,4,5}
diff_set = set1.difference(set2)
print(union_set)
```

```python
# difference alternative
diff_set2 = set1 - set2
print(diff_set2)
```

```python
# Symmetric difference - element in either set but not in both
set1 = {1,2,3}
set2 = {3,4,5,6}
sdiff_set = set1.symmetric_difference(set2)
print(sdiff_set)
```

```python
# symmetric difference alternative
sdiff_set2 = set1 ^ set2
print(sdiff_set2)
```
