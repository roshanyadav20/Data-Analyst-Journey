Basic Data Type in Python

```python
#1. Numeric
a1 = 1
print(type(a1))
a2 = 1.5
print(type(a2))
a3 = complex(2,3)
print(type(a3))
```

```python
#2. Sequence
b1 = "Roshan"
b11 = "123"
print(type(b1))
print(type(b11))
```

```python
b2 = [1,4,3, 'Madhav']
print(type(b2))
```

```python
b2 = (1,4,3, 'Madhav')
print(type(b2))
```

```python
#3. Dictionary
my_dictionary = {'name': 'Roshan', 'age': 22, 'city': 'Sultanpur'}
print(type(my_dictionary))
```

```python
#Sets
my_sets = {1,4,7,'Madhav'}
print(type(my_sets))
```

```python
#5. Boolean
bool1 = True
print(type(bool1))
```

```python
#6. Binary (kam use hota hai)
# bytes, bytearray, memoryview

byte1 = b"Madhav"
print(type(byte1))
```
