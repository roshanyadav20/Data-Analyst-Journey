Loops in Python -> while & for loop

1. While Loop

```python
count = 0
while count < 5: #condition
    print(count)
    count = count + 1
```

```python
# print numbers from 1 to 5 using while loop
count = 1
while count < 6:
    print(count)
    count = count + 1
```

```python
count = 1
while count < 6:
    print(count)
    count += 1
```

```python
count = 5
while count > 0:
    print(count)
    count -= 1
```

```python
count = 5
while count > 0:
    print(count)
    count -= 1
else:
    print("while loop ended")    
```

2. For Loop

```python
language = 'Python' # sequence
for x in language:
    print(x)
```

range() function

range(stop)
range(start, stop, step)

```python
for i in range(5): # start argument
    print(i)
```

```python
for i in range(5, 10): # start stop argument
    print(i)
```

```python
for i in range(1,10,2): # start stop step
    print(i)
```

```python
for i in range(1,10,3):
    print(i)
```

```python
for i in range(5):
    print(i)
else:
    print("for loop ended")    
```

3. Loop Control Statement

- pass statement
- break statement
- continue statement

```python
#1. pass statement - error nhi ayega
for i in range(5):
    # abhi man nhi hai kuch likhne ka
    pass
    
```

```python
count = 5
while count > 0:
    if count == 3:
        pass
    else:
        print(count)
    count -= 1        
```

```python
#2. break statement
for i in range(5):
    if i == 3:
        break
    print(i) # output: 0 1 2
    
```

```python
#3. Continue Statement: ye pura iteratin skip kr deta hai. 
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
while True:
    user_input = input("Enter 'exit' to STOP: ")
    if user_input == 'exit':
        print("congarts! You guessed it right!")
        break
    print("sorry, you entered: ", user_input)
```
