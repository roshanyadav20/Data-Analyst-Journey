2. Acess List - Indexing

- Indexing is used to access one element at a specific position in a list.

```python
list1 = [10,20,30,40,50]

print(list1[0]) # first element
print(list1[1]) # second element 
print(list1[-1]) # last element
print(list1[-2]) # second last element 
```

3. List Slicing:

- Slicing is used to access multiple elements from a list or string.

- Syntax:- list_name[start : stop : step]

- start is inclusive, default value is 0

- stop is exclusive, default is -1/last index value

```python
# first 3 elements
list2 = [10,20,30,40,50,60,100]
print(list2[0:3])
```

```python
# elements from index 1 to 4
print(list2[1:5])
```

```python
# last 3 elements
print(list2[-3:])
```

```python
print(list2[:])
print(list2[::])
```

```python
#alternative range
print(list2[::2]) # step is 2
```

```python
# reverse list 
print(list2[::-1]) # step is -1
```

5. List Methods: 

- Python provides built-in methods to modify and operate on lists.

- 1. append( ): Adds one element at the end of the list.
- 2. extend( ): Adds multiple elements to the list.
- 3. insert( ): Adds an element at a specific position.
- 4. remove( ): Removes the first occurrence of a value.
- 5. pop( ): Removes and returns an element (last by default).
- 6. clear( ): Removes all elements from the list.
- 7. index( ): Returns the index of a given element.
- 8. count( ): Returns how many times an element appears.
- 9. sort( ): Sorts the list in ascending order.
- 10. reverse( ): Reverses the list order.
- 11. copy( ): Creates a shallow copy of the list.

```python
#1. 
fruits2 = ["apple", "banana"]
fruits2.append("cherry")
print(fruits2)
```

```python
#2. extend()
more_fruits = ["orange", "mango"]
fruits2.extend(more_fruits)
print(fruits2)
 
```

```python
#3. insert
fruits3 = ["apple", "banana"]
fruits3.insert(1, "blueberry")
print(fruits3)
```

```python
#4. remove
fruits4 = ["apple", "banana", "orange", "banana"]
fruits4.remove("banana")
print(fruits4)
```

```python
#5. clear
fruits5 = ["apple", "orange"]
fruits5.clear() # empty list
print(fruits5)
```

```python
#6. finding index
fruits = ["apple", "banana", "cherry", "banana"]
index = fruits.index("banana")
print(index)
```

```python
#7 finding index -  within in range
index = fruits.index("banana",2)
print(index)
```

```python
#8. count elements
fruits = ["apple", "banana", "cherry", "banana"]
count = fruits.count("banana")
print(count)
```

```python
#9. reverse string 
fruits = ["apple", "banana", "cherry"]
fruits.reverse()
print(fruits)
```

```python
#10. sorting list
numbers = [40, 10, 30, 20]
numbers.sort() # default sort asc order
print(numbers)
```

```python
#11. sorting list in descending order 
numbers.sort(reverse=True)
print(numbers)
```

```python
#12. sorting string in a list
fruits = ["cherry", "apple", "banana"]
fruits.sort() # default by chars asc order
print(fruits)
```

```python
#13. sorting with a key
fruits.sort(key=len, reverse= True) # sort based on len
print(fruits)
```

```python
#14. pop with index value
numbers = [10,20,30,40]
popped = numbers.pop(2)
print(popped)
print(numbers)
```

```python
#15. pop with default
last = numbers.pop()
print(last)
print(numbers)
```

```python
#16. copying list
fruits = ["apple", "banana", "cherry"]
copy_fruits = fruits.copy() # shallow copy
print(copy_fruits)
```

```python
# copying list - modifying the copy does not effect the original list
copy_fruits.append("date")
print(copy_fruits)
print(fruits)
```

6. Join Lists:

- There are several ways to join, or concatenate, two or more lists in Python.

```python
# Join Lists
list1 = [1,2,3]
list2 = ['a', 'b']

# using + operator
final_list = list1+ list2
print(final_list)
```

```python
# using append method 
list1 = [1,2,3]
list2 = ['a', 'b']
for x in list2:
    list1.append(x)
print(list1)    
    
    
```

```python
# using extend method
list1.extend(list2)
print(list1)
```

7. List Comprehensions:

Syntax:- 

- new_list = [expression for item in iterable if condition]

```python
 #1. create a list of squares
 
squares = [x**2 for x in range(1,6)]
print(squares) 
```

```python
#2. filter even numbers
even_list = [x for x in range(1,10) if x%2 ==0]
print(even_list)
```

```python
#3. apply function to each element of a list
my_list = ['apple', 'mango', 'cherry']

my_name = "madhav"
print(my_name)
print(my_name.upper())

```

```python
my_list = ['apple', 'mango', 'cherry']
print(my_list)
# print(my_list.upper()) # this is wrong way.

uppercase_list = [lst.upper() for lst in my_list]
print(uppercase_list)
```

```python
# flatten a nested list usinglist comprehensions
nested_list = [[1,2],[3,4],[5,6]]

result = [item for sublist in nested_list for item in sublist]
print(result)

# first, [1,2] > 1,2
# second, [3,4] -> 3,4
# third, [5,6] -> 5,6


def flatten_list(lst):
    return [item for sublist in lst for item in sublist]

final_list = flatten_list(nested_list)
print(final_list)
```

8. Iterating Over Lists

```python
# List iterations
fruits = ['apple', 'mango', 'cherry']

# using for loop
for fruit in fruits:
    print(fruit)
    
print("length of list:", len(fruits))    
```

```python
#while loop

index = 0
while index < len(fruits):
    print(fruits[index])
    index += 1
    
print("length of list:", len(fruits))       
```
