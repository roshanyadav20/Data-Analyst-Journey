Operators in Python

```python
 # 1. Arithemetic Operators

 a = 5
 b = 3
 print(a+b) # addition operator 
 print(a-b) # substraction 
 print(a*b) # multiplication 
 print(a/b) # modulus 
 
```

```python
# 2. Comparison Operators - output is boolean value (T/F)

a = 5
b = 3
print(a>b) #grater than operator
print(a<b) #less than
print(a == b) # equal operator
print(a != b) #not equal operator
```

```python
# 3. Assignment Operator

a = 5 # assignment
```

```python
# 4. Logical Operators
# Rule:-
# -> True + True = True
# -> True + False = False
# -> False + False = False

a = 10
b = 20
print(a>10 and b<10) # and operator
```

```python
# and operator
print(a==10 and b ==20)
```

```python
# or operator
print(a==10 or b<10)
```

```python
# 'not' operator
print(not(a==10 and b==20))
```

```python
# 5.Identity operators - is, is not
x = [1,2,3]
y = x
z = [1,2,3]
print(x is y)
print(x is z)

print(x is not z)
```

```python
# 6. Membership operator - in, not in
my_list = ['apple', 'orange', 'papaya']
print('apple' in my_list)
print('apple2' in my_list)
print('apple2' not in my_list) # not in operator
```

```python
# 7. Bitwise Operator- &, |, NOT ~
a = 5  # 5 in binary- 0101
b = 3  # 3 in binary- 0011
print(a & b) # 1 in binary- 0001
```
