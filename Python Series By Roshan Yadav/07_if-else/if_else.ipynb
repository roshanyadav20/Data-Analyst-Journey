1. if statement:-

```python
# 1. if statement - works only for true condition.

a = 26
b = 108
if b>a:
    print("b is greater than a")
```

```python
age = 26
if age > 19:
    print("you are an adult")
```

2. if-else statement

```python
# else handles false condition.
age = int(input("Enter your age: "))
if age > 19:
    print("You are an adult")
else:
    print("You are not an adult")
        
```

```python
temp = 30
if temp < 25:
    print("It's a cool day")
else:
    print("It's a hot day")    
```

3. if-elif-else condition

```python
# multiple conditions lagane ke liye use 
marks = int(input("Enter your marks-100"))
if marks >= 90:
    print("Grade: A+")
elif marks >= 80:
    print("Grade: A")    
elif marks >= 70:
    print("Grade: B")
else: 
    print("Grade: C")   
```

4. Nested if-else statement

```python
# if-else inside if-else statement
# multiple condition depend on each other

# Q: posititve , negative & zero. if num posititve batao - even/odd hai.
number = int(input("Enter a number: "))
if number > 0: #checking posiitvew number
    if number % 2 == 0:
        print("This is a even number")
    else:
        print("This is a odd number")    
else:
    if number == 0:
        print("This is zero")
    else:
        print("This is a negative number")        
```

5. Conditional Expressions (ternary operator)

```python
# Syntax:-
# value_if_true if condititon else value_if_false

age = 26
status = "Major" if age >= 18 else "Minor"
print(status)

```

```python
age = 16
status = "Major" if age >= 18 else "Minor"
print(status)
```
