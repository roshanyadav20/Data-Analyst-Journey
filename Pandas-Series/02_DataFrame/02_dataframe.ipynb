DataFrame:

- create krne ke liye list aur dictionary ki jarurat hoti hai .

```python
import pandas as pd

# list ka use
l = [1,2,3,4,5,6]
var = pd.DataFrame(l)
print(var)
print(type(var))
```

```python
# dictionary ka use

d = {"a":[1,2,3,4,5],"s":[1,2,3,4,5]}

var1 = pd.DataFrame(d) 

print(type(var))
print(var1)
```

```python
# mughe sirf particular column ki need hai 
d = {"a":[1,2,3,4,5],"s":[1,2,3,4,5],"d":[1,2,3,4,5]}

var1 = pd.DataFrame(d,columns=["a"]) 

print(type(var1))
print(var1)
```

```python
# hm jis column ko chahe uss column ko dekh sakte hai.
d = {"a":[1,2,3,4,5],"s":[1,2,3,4,5],1:[1,2,3,4,5]}

var1 = pd.DataFrame(d,columns=["a",1])

print(var1)
print(type(var1))
```

```python
# change index number
d = {"a":[1,2,3,4,5],"s":[1,2,3,4,5],1:[1,2,3,4,5]}
var1 = pd.DataFrame(d,columns=["a",1],index=["a","s","d","f","g"])

print(var1)
print(type(var1))
```

```python
d = {"a":[1,2,3,4,5],"s":[1,2,3,4,5],1:[1,2,3,4,5]}
var1 = pd.DataFrame(d)
print(var1)
print(type(var1))
```

```python
# ek particular data ko get krna
d = {"a":[1,2,3,14,5],"s":[1,2,3,4,5],1:[1,2,3,4,5]}
var1 = pd.DataFrame(d)
print(var1)
print(var1["a"][3])
print(type(var1))
```

List se data frame create karna

```python
list_1 = [[1,2,3,4,5],[11,12,13,14,15]]

var2 = pd.DataFrame(list_1)
print(var2)
print(type(var2))
```

Series se data frame create karna 

```python
sr = {"s":pd.Series([1,2,3,4]),"r":pd.Series([1,2,3,4])}
var3 = pd.DataFrame(sr)
print(var3)
print(type(var2))
```
