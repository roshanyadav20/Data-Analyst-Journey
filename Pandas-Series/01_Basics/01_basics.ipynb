Series:

- It is defined as a one-dimensional array that is capable of storing various data types.

```python
import pandas as pd

x = [3,4,5,7,8]

var = pd.Series(x)

print(var)
print(type(var))
print(var[2])
```

```python
#1. change index number
x = [3,4,5,6]
var = pd.Series(x,index=['a','s','d','f'])
print(var)
```

```python
#2. change data type 
x = [3,4,5,6]
var = pd.Series(x,index=['a','s','d','f'],dtype="float")
print(var)
```

```python
#3. name parameter
x = [3,4,5,6]
var = pd.Series(x,index=['a','s','d','f'],dtype="float",name="python")
print(var)
```

Note:- pandas ke andar series bna rhe hai, series banate samay ham-

- list pass krwa sakte hai.
- Disctionary pass krwa sakte hai.
- Tuple bhi.

```python
# multidimension data likha but output mein one dimension data dikha, isi liye Series bolte hai.

dic = {"name":['python','c','c++','java'],"por":[12,13,14],"rank":[1,4,3,2]}

# series banana hai 
var1 = pd.Series(dic)
print(var1)
# var1 - isko bhi likh sakte hai  

# data type object ka reason:- mix data ka use karna.     
```

Single Data se bhi series bna sakte hai 

```python
s = pd.Series(12)
print(s)
print(type(s))
```

Ek hi data ke multiple series banana:- pass index value, jitna index pass hoga, utni series banegi

```python
s = pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
```

```python
s1 = pd.Series(12,index=[1,2,3,4,5,6,7])
s2 = pd.Series(12,index=[1,2,3,4])

print(s1+s2)
```
